## Plot the Results

```python
import json
import matplotlib.pyplot as plt
import numpy as np

# Before importing matplotlib, set the font family globally
plt.rcParams['font.family'] = 'CMU Serif'

# Read your JSON data
with open('results/rf_binary_cf_pcf_evaluation_metrics.json', 'r') as file:
    data = json.load(file)

pcs = range(2, 18)  # PCs from 2 to 17
scenarios = ['CF Train, CF Test', 'N Train, N Test', 'CF Train, N Test']

fig, axes = plt.subplots(1, 3, figsize=(8, 5))  # 1 row, 3 columns for the subfigures

# Define colors for the plot elements
precision_color = '#FFBF00'
recall_color = '#318CE7'
line_color = 'black'
alpha = 0.6

# Hatches for different fills in grayscale
precision_hatch = '\\\\'  # diagonal hatching
recall_hatch = '//'  # opposite diagonal hatching

# Plot each metric in the specified order for each scenario
for index, scenario in enumerate(scenarios):
    # Extracting the metrics in reverse order so they display from top to bottom correctly
    precisions = [data[f"PC{i}"][scenario.replace('N', f'N{i}')]["Precision"] for i in pcs][::-1]
    recalls = [data[f"PC{i}"][scenario.replace('N', f'N{i}')]["Recall"] for i in pcs][::-1]
    f1_scores = [data[f"PC{i}"][scenario.replace('N', f'N{i}')]["F1 Score"] for i in pcs][::-1]

    y_pos = np.arange(len(pcs))  # y positions for each PC
    width = 0.25  # width of the bars
    
    # Plot Precision as the first bar with hatching
    axes[index].barh(y_pos + width, precisions, width, hatch=precision_hatch, edgecolor='black', color=precision_color, alpha=alpha, label='Precision')
    # Plot F1 Score as a line plot in the middle
    axes[index].plot(f1_scores, y_pos, 'o-', color=line_color, label='F1 Score')
    # Plot Recall as the second bar with hatching
    axes[index].barh(y_pos - width, recalls, width, hatch=recall_hatch, edgecolor='black', color=recall_color, alpha=alpha, label='Recall')


    axes[index].set_xlabel('Scores')
    axes[index].set_title(scenario, fontsize=10)
    # axes[index].grid(True)  # Enable grid
    
    if index == 0:  # Only show y-axis labels for the first subplot
        axes[index].set_yticks(y_pos)
        axes[index].set_yticklabels([f'PC={i}' for i in pcs[::-1]])  # Label in reverse order
    else:
        axes[index].set_yticklabels([])  # Hide y-axis labels for other subplots
        axes[index].set_yticks([])  # Hide y-axis labels for other subplots

# Create custom legend entries
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor=precision_color, alpha=alpha, edgecolor='black', hatch=precision_hatch, label='Precision'),
                   plt.Line2D([0], [0], marker='o', color='black', label='F1 Score', markersize=5),
                   Patch(facecolor=recall_color, alpha=alpha, edgecolor='black', hatch=recall_hatch, label='Recall')]

# fig.legend(handles=legend_elements, loc='upper left', bbox_to_anchor=(1, 0.5))

# plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust layout to make room for the legend

# Placing the legend below the subfigures and center-aligning it
fig.legend(handles=legend_elements, loc='upper center', bbox_to_anchor=(0.4, -0.005), ncol=len(legend_elements))

# Adjust the layout to make room for the legend below the subplots
plt.tight_layout(rect=[0, 0, 0.8, 0.8])

# Save the figure
plt.savefig('results/rf_binary_cf_pcf_evaluation_metrics.png', dpi=300, bbox_inches='tight')

plt.show()
```
